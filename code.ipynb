### problem-1

```python
n=int(input("enter the number:"))
intervals =[]
for _ in range(n):
    start,end=map(int,input("enter the two numbers:").split())
    intervals.append([start,end])
intervals.sort()
m =[]
for i in intervals:
    if not m or m[-1][1]<i[0]:
        m.append(i)
    else:
        m[-1][1] = max(m[-1][1],i[1])

        
for start, end in m:
    print(start, end)
```

### problem-3

```python
s=input("Enter the letters:").strip()
lastseen= {}
l= 0
maxlen = 0
for i in range(len(s)):
    if s[i] in lastseen and lastseen[s[i]]>=l:
        l = lastseen[s[i]] + 1
    lastseen[s[i]] =i
    maxlen = max(maxlen,i-l+ 1)

print("output:",maxlen)
```

### problem-2

```python
n = int(input("enter the number:"))
l=list(map(int,input("enter the number:").split()))
k = int(input("enter the number:"))
window_sum = sum(l[:k])
max_sum = window_sum
for i in range(k, n):
    window_sum += l[i] - l[i - k]
    max_sum = max(max_sum, window_sum)
print(max_sum)
```
